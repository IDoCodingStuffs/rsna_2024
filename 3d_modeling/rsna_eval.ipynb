```python
import pandas as pd
```

```python
sub = pd.read_csv('../data/rsna-2024-lumbar-spine-degenerative-classification/sample_submission.csv')
sub
```

```python
train_df = pd.read_csv('../data/rsna-2024-lumbar-spine-degenerative-classification/train.csv') 

conditions = list(train_df.columns)
conditions.remove("study_id")
conditions = sorted(conditions)
print(conditions)    
```

```python
gt_df = {"row_id":[], "normal_mild": [], "moderate": [], "severe": []}

for study_id in train_df["study_id"]:
    curr = train_df.loc[train_df.study_id == study_id].iloc[0]
    filter = False
    
    for condition in conditions:
        if not isinstance(curr[condition], str):
            filter = True
    if filter:
        continue
    for condition in conditions:
        row_id = f"{study_id}_{condition}"
        if curr[condition] == "Normal/Mild":
            label = "normal_mild"
        else:
            label = curr[condition].lower()
        
        for key in gt_df:
            if key == "row_id":
                gt_df["row_id"].append(row_id)
            elif key == label:
                gt_df[key].append(1)
            else:
                gt_df[key].append(0)

gt_df = pd.DataFrame.from_dict(gt_df)
gt_df
```

```python
merged_df = pd.merge(sub, gt_df, on="row_id", how="inner", suffixes=["_pred", "_gt"])
merged_df
```

```python
merged_df["study_id"] = merged_df["row_id"].str.split("_").str[0]

merged_df["condition"] = merged_df["row_id"].str.split("_").str[1:-2]
merged_df["condition"] = merged_df["condition"].str.join("_")

merged_df["level"] = merged_df["row_id"].str.split("_").str[-2:]
merged_df["level"] = merged_df["level"].str.join("_")

merged_df
```
